````python
# TBC: wheel speeds controllers
import time
from robomaster.robot import Robot

RPM = 63       # ≈ 0.5 m/s
STOP_TIME = 4 

robot = Robot()
robot.initialize(conn_type="sta", sn="1")

try:
    # 1. Avanza a 0.5 m/s
    robot.chassis.drive_wheels(w1=RPM, w2=RPM, w3=RPM, w4=RPM)
    time.sleep(1)
    robot.chassis.drive_wheels()  # stop

    # 2. Avanza e si ferma dopo 2 m (timeout=4s)
    robot.chassis.drive_wheels(w1=RPM, w2=RPM, w3=RPM, w4=RPM, timeout=STOP_TIME)

    # 3. Laterale verso destra
    robot.chassis.drive_wheels(w1=-RPM, w2=RPM, w3=RPM, w4=-RPM)
    time.sleep(1)
    robot.chassis.drive_wheels()

    # 4. Rotazione CW
    robot.chassis.drive_wheels(w1=-RPM, w2=RPM, w3=-RPM, w4=RPM)
    time.sleep(1)
    robot.chassis.drive_wheels()

    # 5. Sinistra 5 RPM, Destra 6 RPM -> curva a SINISTRA (lato destro più veloce)
    robot.chassis.drive_wheels(w1=6, w2=5, w3=6, w4=5)
    time.sleep(2)
    robot.chassis.drive_wheels()

finally:
    robot.close()
````

### Velocity estimation

The robots estimates its linear and angular velocities.
Interested clients can subscribe to this kind of information by specifying a callback called each time new data is available.

To start getting updates on velocity estimations, call [robot.chassis.sub_velocity](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#robomaster-chassis) 
```python
def sub_velocity(freq: int = Literal[1, 5, 10, 20, 50], 
                 callback: Callable[[tuple[float, float, float, float, float, float], None] | None = None): ...
```
that will publish velocity estimations as tuples of 6 numbers.

To stop getting updates, call [robot.chassis.unsub_velocity](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#robomaster-chassis)
```python
def unsub_velocity(): ...
```

The callback, in this case, receives a tuple $(v_x, v_y, v_z, v_{bx}, v_{by}, v_{bz})$ where x is oriented forwards, y rightwards, and z upwards. The first three the x-y-z components in a world-fixed reference frame, the last three the x-y-z components in the robot's reference frame.

For example, to print the linear speed in the robot frame, we can define a callback like

```python
    def cb(data):
        print(f'forwards: {data[3]}, rightwards: {data[4]}')
```
and start a subscription (at 5Hz) 
```python
    robot.chassis.sub_velocity(freq=5, callback=cb)
```
before stopping it after some time
```python
    # wait some time
    time.sleep(5)
    robot.chassis.unsub_velocity()
```

Data for angular velocities is available through [robot.chassis.sub_imu](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#robomaster-chassis), which also provides a tuple of 6 numbers, the last of which is the clock-wise angular speed in degrees per second.

<div class="alert alert-block alert-info">
<b>Control wheel speeds (2)</b>
    
Modify the controller to verify that the robot is indeed moving at the selected velocities (printing them would be enough).
</div>



````python
# TBC: wheel speeds controllers with velocity checks
import time
from robomaster.robot import Robot

RPM = 63  # ≈ 0.5 m/s

def cb_velocity(data):
    print(f"vx (avanti): {data[3]:.3f} m/s | vy (destra): {data[4]:.3f} m/s")

def cb_imu(data):
    print(f"rotazione CW: {data[5]:.3f} °/s")

robot = Robot()
robot.initialize(conn_type="sta", sn="1")

try:
    # Avvia le subscription
    robot.chassis.sub_velocity(freq=5, callback=cb_velocity)

    # --- Test 1: avanti ---
    print("=== AVANTI ===")
    robot.chassis.drive_wheels(w1=RPM, w2=RPM, w3=RPM, w4=RPM)
    time.sleep(2)
    robot.chassis.drive_wheels()  # stop
    time.sleep(0.5)

    # --- Test 2: laterale destra ---
    print("=== LATERALE DESTRA ===")
    robot.chassis.drive_wheels(w1=-RPM, w2=RPM, w3=RPM, w4=-RPM)
    time.sleep(2)
    robot.chassis.drive_wheels()
    time.sleep(0.5)

    # --- Test 3: rotazione CW (usiamo IMU per la velocità angolare) ---
    print("=== ROTAZIONE CW ===")
    robot.chassis.sub_imu(freq=5, callback=cb_imu)
    robot.chassis.drive_wheels(w1=-RPM, w2=RPM, w3=-RPM, w4=RPM)
    time.sleep(2)
    robot.chassis.drive_wheels()
    robot.chassis.unsub_imu()

    # Ferma le subscription
    robot.chassis.unsub_velocity()

finally:
    robot.close()
````

### Control target poses

## Camera


To be able to capture a frame, we have to start the video stream from the robot camera to the Python client using [robot.camera.start_video_stream](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#module-robomaster.camera):
```python
def start_video_stream(
        display: bool = True, 
        resolution: Literal["360p", "540p", "720p"] = "720p"
    ) -> bool: ...
```

<div class="alert alert-block alert-warning">
<b>Use display=False</b> 
    
Do not display the frame when running in a notebook.
</div>

Received frames are stored in a FIFO queue. We can read the first BGR frame (`strategy="pipeline"`) in the queue or the latest received (`strategy="newest"`) using [robot.camera.read_video_frame](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#module-robomaster.camera):
```python
def read_video_frame(
        timeout: float = 3, 
        strategy: Literal["pipeline", "newest"] = "pipeline"
    ) -> numpy.ndarray | None:
```
The call blocks until a frame is available or the timeout passes, in which case it returns `None`.

We should stop the video stream once done using [robot.camera.stop_video_stream](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#module-robomaster.camera)::

```python
def stop_video_stream()-> bool: ...
```

For instance, to capture 10 frames, we can run:
```python
    frames = []
    robot.camera.start_video_stream(display=False)
    while len(frames) < 10:
        frame = robot.camera.read_video_frame()
        if frame is not None:
            frames.append(frame)
    robot.camera.stop_video_stream()
```

<b>Visualize a frame</b>

Take and visualize an image with `matplotlib.pyplot.imshow`. Make sure the colors are correct.
</div>

````python
#TBC: Visualize a frame
import matplotlib.pyplot as plt
from robomaster.robot import Robot

robot = Robot()
robot.initialize(conn_type="sta", sn="1")

try:
    robot.camera.start_video_stream(display=False, resolution="720p")

    frame = None
    while frame is None:
        frame = robot.camera.read_video_frame(strategy="newest")

    robot.camera.stop_video_stream()

finally:
    robot.close()

# Converti BGR → RGB e mostra
frame_rgb = frame[:, :, ::-1]  # inverti i canali: BGR → RGB

plt.figure(figsize=(10, 6))
plt.imshow(frame_rgb)
plt.axis("off")
plt.title("Camera frame")
plt.show()
````

<div class="alert alert-block alert-info">
<b>Capture an object</b>

Save a set of pictures of an object while the robot orbits around it. Approximately one picture every 5 degrees.  Extra: annotate each picture with the corresponding robot orientation you can get from <code>robot.chassis.sub_attitude</code>.
</div>

````python
#TBC: Capture a frame
import time
import threading
import numpy as np
import matplotlib.pyplot as plt
from robomaster.robot import Robot

# --- stato condiviso tra thread ---
current_yaw = 0.0

def cb_attitude(data):
    global current_yaw
    current_yaw = data[2]  # (pitch, roll, yaw) → yaw in gradi

robot = Robot()
robot.initialize(conn_type="sta", sn="1")

frames = []      # lista di (frame_rgb, yaw)
T = 36           # secondi per un giro completo
omega = 10       # °/s
v_tang = 2 * np.pi * 1.0 / T  # ≈ 0.175 m/s, raggio 1m

try:
    robot.camera.start_video_stream(display=False, resolution="360p")
    robot.chassis.sub_attitude(freq=20, callback=cb_attitude)

    robot.chassis.drive_speed(x=0, y=v_tang, z=omega)

    start = time.time()
    last_angle = -999  # angolo dell'ultima foto scattata

    while time.time() - start < T:
        # scatta una foto ogni 5°
        snapped_angle = round(current_yaw / 5) * 5  # arrotonda al multiplo di 5
        if snapped_angle != last_angle:
            frame = robot.camera.read_video_frame(strategy="newest", timeout=0.5)
            if frame is not None:
                frames.append((frame[:, :, ::-1], current_yaw))  # BGR→RGB
                last_angle = snapped_angle
                print(f"Foto #{len(frames):02d} — yaw: {current_yaw:.1f}°")

    robot.chassis.drive_speed()  # stop

finally:
    robot.chassis.unsub_attitude()
    robot.camera.stop_video_stream()
    robot.close()

# --- visualizza una griglia di frame annotati ---
n = len(frames)
cols = 8
rows = (n + cols - 1) // cols

fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 3))
axes = axes.flatten()

for i, (img, yaw) in enumerate(frames):
    axes[i].imshow(img)
    axes[i].set_title(f"{yaw:.1f}°", fontsize=8)
    axes[i].axis("off")

# nascondi celle vuote
for j in range(i + 1, len(axes)):
    axes[j].axis("off")

plt.suptitle("Orbita 360° — una foto ogni ~5°", fontsize=14)
plt.tight_layout()
plt.show()
````

<div class="alert alert-block alert-info">
<b>Detect a ball</b>

Assume that the ball is the only red object in the world. Take a few pictures of the ball and save them. Design a function that given a picture returns the mask of the pixels belonging to the ball (use thresholds on BGR, or on the hue and saturation components, computed using [skimage.color.rgb2hsv](https://scikit-image.org/docs/stable/api/skimage.color.html#skimage.color.rgb2hsv))
</div>

````python
#TBC: Detect a ball (segmentation)
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2hsv
from robomaster.robot import Robot

# ── 1. Cattura alcune foto ──────────────────────────────────────────────────
robot = Robot()
robot.initialize(conn_type="sta", sn="1")

captured = []
try:
    robot.camera.start_video_stream(display=False, resolution="720p")
    while len(captured) < 5:
        frame = robot.camera.read_video_frame(strategy="newest")
        if frame is not None:
            captured.append(frame[:, :, ::-1].copy())  # BGR → RGB
            time.sleep(0.5)
    robot.camera.stop_video_stream()
finally:
    robot.close()

# ── 2. Funzione di detection ────────────────────────────────────────────────
def detect_ball(img_rgb: np.ndarray) -> np.ndarray:
    """
    Ritorna una maschera booleana True sui pixel appartenenti alla palla rossa.
    Input:  img_rgb  — array (H, W, 3) uint8, valori 0-255
    Output: mask     — array (H, W) bool
    """
    # skimage.rgb2hsv vuole float [0,1]
    img_float = img_rgb.astype(np.float32) / 255.0
    hsv = rgb2hsv(img_float)

    H = hsv[:, :, 0]  # hue        0–1
    S = hsv[:, :, 1]  # saturation 0–1
    V = hsv[:, :, 2]  # value      0–1

    # Rosso: hue vicino a 0 (0–0.05) OPPURE vicino a 1 (0.95–1.0)
    red_low  = (H < 0.05)
    red_high = (H > 0.95)
    hue_mask = red_low | red_high

    # Filtri su saturazione e luminosità per escludere bianco/grigio/nero
    sat_mask = S > 0.45   # colore abbastanza saturo
    val_mask = V > 0.20   # non troppo scuro

    mask = hue_mask & sat_mask & val_mask
    return mask

# ── 3. Visualizza risultati ─────────────────────────────────────────────────
fig, axes = plt.subplots(len(captured), 3, figsize=(12, 4 * len(captured)))

for i, img in enumerate(captured):
    mask = detect_ball(img)

    # immagine originale
    axes[i, 0].imshow(img)
    axes[i, 0].set_title(f"Frame {i+1} — originale")
    axes[i, 0].axis("off")

    # maschera
    axes[i, 1].imshow(mask, cmap="gray")
    axes[i, 1].set_title("Maschera palla")
    axes[i, 1].axis("off")

    # overlay
    overlay = img.copy()
    overlay[~mask] = (overlay[~mask] * 0.3).astype(np.uint8)  # scurisci sfondo
    axes[i, 2].imshow(overlay)
    axes[i, 2].set_title("Overlay")
    axes[i, 2].axis("off")

plt.tight_layout()
plt.show()
````

````python
#TBC: Detect a ball (centroid)
from skimage.measure import label, regionprops

def detect_ball_centroid(img_rgb: np.ndarray) -> dict | None:
    """
    Trova il centroide e la dimensione della palla rossa nell'immagine.

    Input:  img_rgb  — array (H, W, 3) uint8
    Output: dict con 'centroid' (row, col), 'area' (n° pixel), 'bbox'
            oppure None se nessuna componente trovata
    """
    mask = detect_ball(img_rgb)

    # Etichetta le componenti connesse (4-connectivity di default)
    labeled = label(mask)          # ogni componente ha un intero diverso
    regions = regionprops(labeled) # lista di oggetti RegionProperties

    if not regions:
        return None

    # Prendi la componente più grande per area
    largest = max(regions, key=lambda r: r.area)

    return {
        "centroid": largest.centroid,  # (row, col) in pixel
        "area":     largest.area,      # numero di pixel
        "bbox":     largest.bbox,      # (min_row, min_col, max_row, max_col)
    }

import matplotlib.pyplot as plt
import matplotlib.patches as patches

for i, img in enumerate(captured):
    result = detect_ball_centroid(img)

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.imshow(img)

    if result is not None:
        cy, cx = result["centroid"]   # centroid è (row, col) → (y, x)
        area    = result["area"]
        r0, c0, r1, c1 = result["bbox"]

        # Centroide
        ax.plot(cx, cy, "x", color="yellow", markersize=14, markeredgewidth=3)

        # Bounding box
        rect = patches.Rectangle(
            (c0, r0), c1 - c0, r1 - r0,
            linewidth=2, edgecolor="lime", facecolor="none"
        )
        ax.add_patch(rect)
        ax.set_title(f"Frame {i+1} — centroide: ({cx:.1f}, {cy:.1f})  area: {area} px")
    else:
        ax.set_title(f"Frame {i+1} — nessuna palla rilevata")

    ax.axis("off")
    plt.tight_layout()
    plt.show()

````

##  Distance sensor

To get distance readings from the front-facing sensor, we start a subscriber using [robot.sensor.sub_distance](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#module-robomaster.sensor)
```python
def sub_distance(
    freq: int = Literal[1, 5, 10, 20, 50], 
    callback: Callable[[tuple[float, float, float, float], None] | None = None
    ) -> bool : ...
```
which we eventually stop with [robot.sensor.unsub_distance](https://robomaster-dev.readthedocs.io/en/latest/python_sdk/robomaster.html#module-robomaster.sensor)

```python
def unsub_distance(): ...
```

The callback receives a tuple of 4 numbers, the first being the distance read by the sensor in mm.
A value of `65534` denotes no detection within the range of view.

For example, to store the last N readings in a circular buffer, we can use a callback like

```python
readings = queue.Queue(maxsize=50)

def cb(data):
    while readings.full():
        readings.get_nowait()
    readings.put(data[0])

robot.sensor.sub_distance(freq=10, callback=cb)
```

````python
# TBC: plot readings
import time
import queue
import numpy as np
import matplotlib.pyplot as plt
from robomaster.robot import Robot

# --- stato condiviso ---
dist_readings = []   # lista di (yaw_deg, distance_mm)
current_yaw   = 0.0

def cb_attitude(data):
    global current_yaw
    current_yaw = data[2]  # (pitch, roll, yaw) → yaw in gradi

def cb_distance(data):
    d = data[0]
    dist_readings.append((current_yaw, d))

# --- rotazione 360° in 12s → 30°/s ---
OMEGA   = 30    # °/s
T_ROT   = 12    # secondi (>360°/30 = 12 → un giro esatto)

robot = Robot()
robot.initialize(conn_type="sta", sn="1")
try:
    robot.chassis.sub_attitude(freq=50, callback=cb_attitude)
    robot.sensor.sub_distance(freq=50, callback=cb_distance)
    
    robot.chassis.drive_speed(x=0, y=0, z=OMEGA)
    time.sleep(T_ROT)
    robot.chassis.drive_speed()  # stop
    
    robot.sensor.unsub_distance()
    robot.chassis.unsub_attitude()
finally:
    robot.close()

# --- plot ---
angles = np.array([r[0] for r in dist_readings])
dists  = np.array([r[1] for r in dist_readings], dtype=float)

# Sostituisci "no detection" (65534) con NaN per non sporcare il plot
dists[dists >= 65534] = np.nan

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1. Plot cartesiano
axes[0].plot(angles, dists / 1000, color="steelblue", linewidth=1.5)
axes[0].axhline(y=np.nanmean(dists/1000), color="red", linestyle="--", label="media")
axes[0].set_xlabel("Angolo yaw (°)")
axes[0].set_ylabel("Distanza (m)")
axes[0].set_title("Distanza vs angolo di rotazione")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 2. Plot polare — più intuitivo per una rotazione
ax_polar = plt.subplot(122, projection="polar")
theta = np.radians(angles)
ax_polar.plot(theta, dists / 1000, color="steelblue", linewidth=1.5)
ax_polar.set_title("Vista polare (distanza in m)", pad=15)

plt.tight_layout()
plt.savefig("distance_rotation.png", dpi=150)
plt.show()

print(f"Totale campioni: {len(dist_readings)}")
print(f"Distanza min:  {np.nanmin(dists)/1000:.3f} m")
print(f"Distanza max:  {np.nanmax(dists)/1000:.3f} m")
print(f"Distanza media: {np.nanmean(dists)/1000:.3f} m")
````

<div class="alert alert-block alert-info">
<b>Find the sensor range</b>

Design and perform an experiment to find the maximal range of the sensor.
</div> 

````python
# TBC: max range
import time
import queue
import numpy as np
import matplotlib.pyplot as plt
from robomaster.robot import Robot

STEP_M     = 0.20   # avanza 20 cm per volta
MAX_STEPS  = 25     # fino a 5 metri totali
SPEED      = 0.3    # m/s durante lo spostamento
N_SAMPLES  = 15     # campioni per misura stazionaria

NO_DETECT  = 65534

dist_queue = queue.Queue(maxsize=50)

def cb_distance(data):
    while dist_queue.full():
        dist_queue.get_nowait()
    dist_queue.put(data[0])

def flush_queue():
    while not dist_queue.empty():
        dist_queue.get_nowait()

def get_stable_reading(n=N_SAMPLES, wait=1.5):
    """Aspetta n campioni e ritorna media e std (ignora 65534)."""
    flush_queue()
    time.sleep(wait)
    samples = []
    while not dist_queue.empty():
        v = dist_queue.get_nowait()
        if v < NO_DETECT:
            samples.append(v)
    if not samples:
        return None, None
    return np.mean(samples), np.std(samples)

# --- esperimento ---
results = []   # lista di (estimated_robot_pos_m, mean_dist_mm, std_mm, detected)
position = 0.0

robot = Robot()
robot.initialize(conn_type="sta", sn="1")
try:
    robot.sensor.sub_distance(freq=50, callback=cb_distance)
    time.sleep(0.5)

    for step in range(MAX_STEPS):
        print(f"\n--- Step {step+1} | posizione robot: {position:.2f} m ---")

        # misura stazionaria
        mean_d, std_d = get_stable_reading()

        if mean_d is not None:
            detected = True
            print(f"  Distanza: {mean_d/1000:.3f} m ± {std_d/1000:.4f} m")
        else:
            detected = False
            print(f"  NO DETECTION (>range)")

        results.append((position, mean_d, std_d, detected))

        # se tre step consecutivi non rilevano → abbiamo superato il range
        last3_detected = [r[3] for r in results[-3:]]
        if len(last3_detected) == 3 and not any(last3_detected):
            print("\n>>> Range massimo superato — fermo il robot.")
            break

        # avanza di un STEP verso la parete
        t_step = STEP_M / SPEED
        robot.chassis.drive_speed(x=-SPEED, y=0, z=0)  # indietro = allontana dalla parete
        time.sleep(t_step)
        robot.chassis.drive_speed()
        position += STEP_M
        time.sleep(0.3)

    robot.sensor.unsub_distance()
finally:
    robot.close()

# --- analisi risultati ---
positions  = np.array([r[0] for r in results])
mean_dists = np.array([r[1] if r[1] is not None else np.nan for r in results])
std_dists  = np.array([r[2] if r[2] is not None else 0.0   for r in results])
detected   = np.array([r[3] for r in results])

# distanza totale sensore-parete = dist_iniziale - posizione_robot
# (il robot si avvicina, quindi la distanza letta dovrebbe scendere)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Plot 1: distanza letta vs posizione robot ---
axes[0].errorbar(
    positions[detected], mean_dists[detected] / 1000,
    yerr=std_dists[detected] / 1000,
    fmt="o-", color="steelblue", capsize=4, label="Rilevato"
)
axes[0].scatter(
    positions[~detected], np.zeros(np.sum(~detected)),
    marker="x", color="red", s=80, zorder=5, label="No detection"
)
axes[0].axvline(
    x=positions[~detected][0] if np.any(~detected) else positions[-1],
    color="red", linestyle="--", label="Soglia range"
)
axes[0].set_xlabel("Distanza robot-partenza (m)")
axes[0].set_ylabel("Distanza letta dal sensore (m)")
axes[0].set_title("Caratterizzazione range sensore")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# --- Plot 2: distanza letta vs distanza attesa (linearità) ---
# distanza attesa = mean_dists[step=0] - position (robot si avvicina)
if mean_dists[0] is not np.nan:
    expected = mean_dists[0] - positions * 1000  # in mm
    axes[1].scatter(
        expected[detected] / 1000,
        mean_dists[detected] / 1000,
        color="steelblue", label="Letture"
    )
    lim = max(np.nanmax(expected), np.nanmax(mean_dists)) / 1000
    axes[1].plot([0, lim], [0, lim], "r--", label="Ideale (y=x)")
    axes[1].set_xlabel("Distanza attesa (m)")
    axes[1].set_ylabel("Distanza letta (m)")
    axes[1].set_title("Linearità del sensore")
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("sensor_range_experiment.png", dpi=150)
plt.show()

# --- Stampa risultato finale ---
if np.any(~detected):
    max_range = positions[~detected][0]
    print(f"\n✅ Range massimo stimato: {max_range:.2f} m")
else:
    print(f"\n⚠️  Sensore ha sempre rilevato — aumenta MAX_STEPS o STEP_M")
````

<b>Turn towards near object</b>

Write a simple controller such that: the robot slowly rotates in place. As soon as it sees an object that is less than 1m far, it lights up red and stops.
</div> 

````python
# TBC: turn towards near object
import time
from robomaster.robot import Robot

THRESHOLD_MM = 1000   # 1 metro
OMEGA        = 20     # °/s rotazione lenta

robot = Robot()
robot.initialize(conn_type="sta", sn="1")
try:
    latest_dist = [float("inf")]  # lista per mutabilità nel callback

    def cb_distance(data):
        latest_dist[0] = data[0]

    robot.sensor.sub_distance(freq=20, callback=cb_distance)
    time.sleep(0.3)  # aspetta i primi campioni

    robot.chassis.drive_speed(x=0, y=0, z=OMEGA)  # inizia a ruotare

    while True:
        d = latest_dist[0]
        if d < THRESHOLD_MM:          # oggetto rilevato entro 1m
            robot.chassis.drive_speed()                            # stop
            robot.led.set_led(comp="all", r=255, g=0, b=0, effect="on")
            print(f"Oggetto rilevato a {d/1000:.3f} m — fermo!")
            break
        time.sleep(0.05)

    robot.sensor.unsub_distance()
finally:
    robot.close()


MAX_WAIT = 30  # secondi
start = time.time()

robot.chassis.drive_speed(x=0, y=0, z=OMEGA)

while time.time() - start < MAX_WAIT:
    if latest_dist[0] < THRESHOLD_MM:
        robot.chassis.drive_speed()
        robot.led.set_led(comp="all", r=255, g=0, b=0, effect="on")
        print(f"Oggetto a {latest_dist[0]/1000:.3f} m")
        break
    time.sleep(0.05)
else:
    robot.chassis.drive_speed()
    robot.led.set_led(comp="all", r=0, g=255, b=0, effect="on")  # verde = nessun oggetto
    print("Timeout — nessun oggetto trovato entro 1m")
````
